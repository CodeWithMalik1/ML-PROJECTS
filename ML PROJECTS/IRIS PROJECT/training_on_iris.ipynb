```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
```

```python
data=pd.read_excel("train_data1.xlsx")
print(data.head())
```

```python
data.dropna()
```

```python
data.isnull().sum()
```

```python
x=data.iloc[:,:-1]
```

```python
x
```

```python
y=data.iloc[:,-1]
y
```

```python
model = RandomForestClassifier()
```

```python
x.values
```

```python
model.fit(x.values, y.values)
```

```python
predictions = model.predict([[4.4,2.9,1.4,0.2]])  # Predict on the same/new data (for demonstration)
print(predictions)  # Display predictions
```

```python
predictions
```

```python
test_dataset=pd.read_excel("test_dataa.xlsx")
x_test=test_dataset.iloc[:, :-1]
predictions_of_test_data=model.predict(x_test.values)
```
